# クラスタリング

教師なし学習

```
from sklearn.datasets import make_blobs
```

```
X, y = make_blobs(n_samples=200, centers=4, n_features=2, cluster_std=2.0, random_state=42)
```

```
%matplotlib inline
from matplotlib import pyplot
pyplot.rcParams['font.family'] = 'IPAPGothic'  # 日本語フォントの設定
```

```
pyplot.figure(figsize=(12, 12))

# データの散布図をプロットする
pyplot.scatter(X[:, 0], X[:, 1], s=100)

pyplot.xlabel(f"$x_{0}$", fontsize=16)
pyplot.ylabel(f"$x_{1}$", fontsize=16)
pyplot.grid()
pyplot.show()
```

```
# クラスタ毎にマーカーを変える
# https://matplotlib.org/api/markers_api.html#module-matplotlib.markers
markers = ["o", "v", "s", "P"]

pyplot.figure(figsize=(12, 12))

# 今回は4つのクラスタを作成した
# 各クラスタについて、
for i in range(4):
    idx = (y == i)  # クラスタ番号が一致するインデックス
    
    # 散布図をプロットする
    pyplot.scatter(X[idx, 0],
                   X[idx, 1],
                   s=100,
                   marker=markers[i],
                   label=f"クラスタ{i}")
    
pyplot.xlabel(f"$x_{0}$", fontsize=16)
pyplot.ylabel(f"$x_{1}$", fontsize=16)

pyplot.grid()
pyplot.legend()
pyplot.show()
```
